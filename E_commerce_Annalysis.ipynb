```python
#TRAINING THE MODEL
```

```python
# Define X (features) and y (target)
X = df.drop(columns=['Churn'],axis=1)  # the  features minus churn
y = df['Churn']# this is the column you are trying to predict if curn or not

```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

```python
#test and predict by creating a model: from the linear model family, we will import logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#If your features are not scaled, the model might struggle to converge. Try normalizing your data using StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#We create an instance of the Logistic Regression
logmodel = LogisticRegression(max_iter=1000) #Try increasing the maximum number of iterations so the model has more time to converge:(1000)
#LET'S TRAIN THE MODEL
logmodel.fit(X_train_scaled, y_train)
print("Training Accuracy:", logmodel.score(X_train_scaled, y_train))
print("Testing Accuracy:", logmodel.score(X_test_scaled, y_test))
```

```python
#OR YOU CAN RUN THIS
```

```python
logmodel = LogisticRegression(solver='saga', max_iter=1000)
logmodel.fit(X_train_scaled, y_train)
print("Training Accuracy:", logmodel.score(X_train_scaled, y_train))
print("Testing Accuracy:", logmodel.score(X_test_scaled, y_test))
```

```python
#OR
```

```python
logmodel = LogisticRegression(class_weight='balanced', max_iter=1000, solver='saga')
logmodel.fit(X_train_scaled, y_train)
print("Training Accuracy:", logmodel.score(X_train_scaled, y_train))
print("Testing Accuracy:", logmodel.score(X_test_scaled, y_test))
```

```python
y_probs = logmodel.predict_proba(X_test_scaled)[:,1]  # Get probability of Class 1
threshold = 0.3  # Adjust this value
y_pred_adjusted = (y_probs >= threshold).astype(int)

print(classification_report(y_test, y_pred_adjusted))
```

Accuracy: 69% (low)
Precision (Class 1 - Churn): 34% (low)
Recall (Class 1 - Churn): 92% (high) ✅
F1-score (Class 1 - Churn): 50% (moderate)
🔍 What this means:

The model detects almost all churn cases (high recall, 92%), but it also misclassifies many non-churn customers as churn (low precision, 34%).
Good if you want to catch all possible churn cases, but many false positives might cause unnecessary intervention.


 Solution 2 (Using a Different Model)

🔍 What this means:

The model accurately identifies churn cases with fewer false positives (high precision, 96%).
It still catches most churn cases (recall 78%), though slightly lower than Solution 2.

- Which One is Better?
It depends on your goal:

If you want to detect as many churn customers as possible (high recall), use Solution 1.
If you want a balanced model that detects churn without too many false positives, Solution 2 is the best.
 Final Recommendation:
- Solution 2 would be the best choice overall because it has high accuracy (96%), strong precision (96%), and good recall (78%)—which means it's both reliable and efficient. But, since the goal is to identify customers likely to churn and take proactive measures (e.g., promotions, retention offers), the best solution should prioritize recall (catching as many churners as possible), which is Solution 1
